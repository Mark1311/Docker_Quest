# Docker

# What is Docker?

Docker ek platform hai jo developers ko software applications ko container mein package karne, distribute karne aur run karne mein madad karta hai. Container ek lightweight aur isolated environment hota hai jisme aap apni applications aur unki dependencies ko ek saath bundle kar sakte ho. Iska faida ye hai ki jo software ek system pe properly kaam karta hai, wo doosre systems pe bhi same tarike se kaam karega, chahe wo development machine ho, testing environment ho ya production server ho.

### Docker ke kuch key features:

Containers: Ye applications ko lightweight aur isolated manner mein run karte hain.

Portability: Docker container ko kahin bhi run kiya ja sakta hai, chahe wo developer ka machine ho ya cloud environment.

Scalability: Docker ko easily scale kiya ja sakta hai jab zarurat ho.

Efficiency: Docker resources ko efficiently use karta hai, jaise ki CPU aur memory.

Docker se development aur deployment ka process kaafi fast aur efficient ho jata hai, jo team collaboration ko bhi asaan banata hai.

# Docker Kya Hai?

Docker ek open-source platform hai jo containerization technology par kaam karta hai. Iska main purpose hai software applications ko ek standardized format mein package karna, taaki woh har environment mein easily chal sakein. Matlab, aap jo application apne local machine par run kar rahe hain, woh bilkul wahi application aapke production server pe bhi run kar sakta hai, bina kisi issue ke.

## Containerization Kya Hai?

Container ek lightweight aur isolated environment hai. Aap ismein apne application ko run karte ho, uske saath saath uski dependencies bhi include kar lete ho (jaise ki libraries, packages, configurations, etc.). Iska matlab hai ki aapko har bar naye environment ko setup karne ki zarurat nahi padti, aur aapka software har jaga same tarike se kaam karta hai.

## Docker Ke Components:

### Docker Engine:

Ye ek client-server architecture hai jo Docker containers ko manage karta hai. Server part ko Docker Daemon kaha jata hai, jo containers ko create, run aur manage karta hai. Client part, jo command line interface (CLI) ke through hota hai, user ko Docker Daemon se interact karne ka tareeka deta hai.

### Images:

Docker images ek tarah se blueprints hoti hain jo container banane ke liye use hoti hain. Image mein application aur uski dependencies included hoti hain. Ye immutable hoti hain, iska matlab hai ki ek baar image ban gayi, toh usko badla nahi ja sakta.
Aap Docker Hub se publicly available images download kar sakte hain ya khud bhi apni custom images create kar sakte hain.

### Containers:

Jab aap ek image ko run karte hain, toh ek container ban jata hai. Container ek isolated environment hai jisme aapka application chal raha hota hai. Yeh ek lightweight process hai jo apne resources (CPU, memory, etc.) ko use karta hai lekin woh host machine se isolated hota hai.

### Dockerfile:

Dockerfile ek script hoti hai jisme aap define karte hain ki aapki image kaise banegi. Ismein aap apni application ki dependencies, configuration, aur other setup instructions likhte hain.

### Docker Hub:

Docker Hub ek public repository hai jahan pe aap Docker images ko upload ya download kar sakte ho. Bahut si ready-made images, jaise ki Node.js, MySQL, Python, etc., aap Docker Hub se easily access kar sakte hain.
Docker Ke Benefits:

### Portability:

Docker containers ek jagah se doosri jagah pe move kiye ja sakte hain bina kisi configuration change ke. Matlab, agar aapka application local machine pe chal raha hai, toh woh production server pe bhi bina kisi issue ke chal sakta hai.

### Isolation:

Har container apne environment mein kaam karta hai, isliye ek application ka bug ya issue doosre application ko affect nahi karta. Yeh aapko flexibility deta hai.

### Consistency:

Agar ek application ek container mein run kar raha hai, toh woh same application har jagah same tarike se chalega. Yeh development aur deployment ke beech consistency banata hai.

### Scalability:

Docker aapko easily application ko scale karne ki suvidha deta hai. Aap containers ko quickly replicate kar sakte hain jab load badhta hai.

### Efficiency:

Containers resource-efficient hote hain, kyunki ye virtual machines (VMs) ke comparison mein kam memory aur CPU use karte hain. Containers ek hi OS kernel share karte hain, jabki VMs apne apne separate operating systems chalte hain.

## Docker Workflow Example:

### Dockerfile likhna:
    
Aap apne project ke root directory mein ek Dockerfile banate hain. Ismein aap specify karte hain ki kaunsa base image use karna hai, kis tarah ki dependencies install karni hain, aur kaunsa command run karna hai.

### Image Build Karna:

docker build -t myapp . command ke through aap apni image build karte hain.

### Container Run Karna:

Once image ban gayi, us image se ek container run karne ke liye aap docker run command ka use karte hain.

### Container Management:

Aap containers ko manage karne ke liye docker ps, docker stop, docker start, etc. commands use kar sakte hain.

### Docker Ka Use Cases:

Microservices Architecture: Docker ko microservices ko deploy karne ke liye extensively use kiya jata hai, jahan har service apne container mein hoti hai.
Continuous Integration (CI) & Continuous Deployment (CD): Docker CI/CD pipelines mein bhi use hota hai taaki code changes ko test aur deploy kiya ja sake bina environment specific issues ke.
Testing: Docker containers mein aap apne application ko easily test kar sakte hain.

# Why we used Docker

Docker is used to simplify the process of developing, testing, and deploying applications by creating lightweight, consistent, and isolated environments known as containers. Here's why Docker is so popular and useful:

## Portability: 

Docker containers can run on any machine that has Docker installed, regardless of the underlying operating system or hardware. This ensures that the application behaves the same in development, testing, and production environments.

## Isolation: 
    
Docker allows you to isolate different applications or services in separate containers. Each container has its own dependencies, configuration, and environment, reducing the risk of conflicts between them.

## Efficiency: 

Docker containers are lightweight compared to virtual machines because they share the host system's kernel instead of running a full OS. This leads to faster startup times and better resource utilization.

## Consistency: 

By using Docker, you can package all dependencies (libraries, configurations, etc.) within a container, ensuring that the app works the same way regardless of where it runs. This eliminates the common issue of "it works on my machine."

## Scalability: 

Docker makes it easier to scale applications, especially when combined with orchestration tools like Kubernetes. You can quickly deploy, manage, and scale containers across multiple machines or cloud environments.

## Version Control: 

Docker images (the templates for creating containers) can be versioned and stored in repositories like Docker Hub. This allows teams to track changes, rollback to previous versions, and manage dependencies more effectively.

## Continuous Integration/Continuous Deployment (CI/CD): 
                                                                                                                                                     Docker is a great fit for modern CI/CD pipelines, where applications need to be built, tested, and deployed automatically in a consistent environment.

# Image 

Docker Image kya hoti hai? Docker image ek tarah ka snapshot hota hai jo aapke application ka poora environment (code, dependencies, libraries, configuration) ko encapsulate karta hai. Iska mtlb hai ki jab aap Docker image create karte ho, toh aap apne application ko ek portable aur self-contained unit mein pack karte ho. Aap is image ko kisi bhi system pe run kar sakte ho, jo Docker ko support karta ho.

### Docker Image ka Structure:

## Base Image: 

Docker image ka base ek operating system ya software environment hota hai. For example, agar aap Python application bana rahe ho, toh aap ek Python base image le sakte ho.

## Layers:

Docker images ko multiple layers mein divide kiya jata hai. Har layer ek specific instruction ko represent karti hai, jaise ki ek file ko copy karna, dependencies install karna, etc. Layers ka concept image ko efficient banaata hai kyunki Docker layers ko cache karta hai aur agar same layer pe koi changes nahi hoti toh wo dobara build nahi hoti.

## Dockerfile: 

Docker image ko build karne ke liye aapko ek file ki zarurat hoti hai jise Dockerfile kaha jata hai. Dockerfile mein instructions hoti hain jo define karti hain ki image ko kaise build karna hai. Jaise ki, kaunsa base image use karna hai, kaunse commands run karne hain, kaunse files copy karni hain, etc.

##### Docker image mein typically yeh cheezein hoti hain:

Operating System: Base OS ya uska ek minimal version, jisme application run kar sake.

Dependencies: Software ko chalane ke liye zaroori libraries aur tools.

Application Code: Wo code jo aapne likha hai, jise run karna hai.

Configuration: Settings ya configuration files jo software ko customize karti hain.

# Containerization vs Virtualization:

### Virtualization: 

Ye ek purani technique hai jisme ek physical machine ke andar multiple virtual machines (VMs) run ki jati hain. Har VM apni full operating system (OS) aur kernel ke saath chalti hai.
Isme resource overhead zyada hota hai, kyunki har VM ko apni OS chalani padti hai.

### Containerization: 

Docker containers ek shared OS kernel ka use karte hain, isliye ye lightweight aur fast hote hain. Ek container apni application ko run karta hai, lekin ye directly underlying OS ke kernel ka use karta hai bina apna alag OS install kiye.
Isme resource overhead kam hota hai, aur containers bahut jaldi start hote hain.


# Docker Architecture:
    
Docker ka architecture mainly 4 key components se milke bana hota hai:

## Docker Daemon (dockerd):

Ye background mein run karta hai aur Docker containers ko manage karta hai.

Docker daemon apne container ko create, start, stop, aur delete karne ke liye request handle karta hai.
    
Docker Daemon images ko build karne, containers ko execute karne, aur network manage karne ka kaam karta hai.

## Docker Client:

Ye ek command-line tool hota hai, jise aap terminal pe use karte hain. Jab aap docker run ya docker build command chalate hain, to ye Docker Client se communicate karta hai.

Docker client actually Docker Daemon ke sath interact karta hai aur user commands ko process karta hai.

## Docker Images:

Docker images ek read-only template hoti hain jisme application ke code, libraries, aur dependencies hote hain.

Images ko ek Dockerfile se create kiya jata hai, jo ek configuration file hota hai jisme instructions hoti hain ki image kaise banani hai.

Docker images ek tarah se blueprint hoti hain, jisse aap containers create karte ho.

## Docker Containers:

Docker containers ek chhota, isolated environment hote hain jo image ke andar defined application ko run karte hain.
Har container ek running instance hota hai ek image ka.
Container apne environment ko isolated rakhta hai, isliye agar ek container crash ho jata hai to dusre containers ko usse koi farak nahi padta.

## Docker Registry (Docker Hub):

Docker Hub ek public registry hai jahan par aap Docker images ko store kar sakte hain, share kar sakte hain, aur download kar sakte hain.
Docker Hub mein aapko kai predefined images milti hain, jaise ki Python, Node.js, Nginx, etc.

# Advantage of Docker?

## Portability:

## Isolation: 

Docker containers apne dependencies aur environment se isolated hote hain. Matlab, agar aapka ek application container mein hai, to doosra application ya environment uspe asar nahi dalega. Is se aapke applications safe rehte hain aur conflicts kam ho jaate hain.

## Scalability: 

Docker ko easily scale kiya jaa sakta hai. Agar aapko zyada resources ki zarurat ho, to aap containers ko horizontally scale kar sakte hain, yani zyada containers chala sakte hain without much effort.

## Efficiency: 

Docker ka overhead kaafi kam hota hai compared to virtual machines. Containers ko start karna aur stop karna bhi fast hota hai, jisse aapke resources efficiently use hote hain.

## Version Control aur Rollbacks: 

Docker images ko version control ke through manage kiya jaa sakta hai, aur agar koi version issue ho to aap easily previous version pe rollback kar sakte hain.

## Easy to use CI/CD Integration:

Docker ko continuous integration (CI) aur continuous deployment (CD) processes ke saath integrate kiya jaa sakta hai, jo software development ke process ko automate karne mein madad karta hai.

## Cost Efficiency: 

Docker containers lightweight hote hain, jo aapko apne infrastructure ko optimize karne mein madad karte hain. Isse hardware resource utilization behtar hota hai aur overall cost bhi kam ho sakti hai.

## Ecosystem and Community Support: 

Docker ka ek bohot bada ecosystem aur community hai, jisme aapko tools, libraries aur documentation milti hain jo development aur troubleshooting mein madad karti hain.

# Docker and VM?

## Architecture:

#### Virtual Machine (VM): 

VM ek full operating system ko virtualize karta hai. Isme ek hypervisor hota hai jo multiple virtual machines ko ek physical machine par run karne deta hai. Har VM apna apna OS (Windows, Linux, etc.) run karta hai, jo heavy resource consumption ka sabab ban sakta hai.

#### Docker:
    
Docker ek containerization tool hai, jo applications ko containers ke andar run karta hai. Containers ek shared OS kernel ka use karte hain, lekin har container apna independent user space hota hai. Ismein operating system ka full virtualisation nahi hota, jo zyada efficient hota hai.

## Resource Usage:

#### VM:
    
VM har bar apna operating system load karta hai, jiska matlab hai ke zyada resources (CPU, memory) ki zarurat hoti hai.

#### Docker: 

Docker containers shared OS kernel ka use karte hain, isliye resources ka consumption kaafi kam hota hai. Docker lightweight hota hai aur har container ko quickly start/stop kiya ja sakta hai.

## Performance:

#### VM: 

VM ki performance thodi kam ho sakti hai, kyunki usme full OS ko run karna padta hai, jisme extra overhead hota hai.

#### Docker: 

Docker containers ki performance better hoti hai, kyunki yeh OS ke kernel ko directly access karte hain aur inme overhead kaafi kam hota hai.

## Isolation:

#### VM: 

VM ek completely isolated environment provide karta hai, jisme different operating systems run karte hain. Har VM apna ek OS, libraries, aur binaries leke aata hai.

#### Docker: 

Docker containers mein bhi isolation hota hai, lekin yeh OS kernel ko share karte hain, isliye thoda less isolated hote hain as compared to VMs.

## Use Cases:

#### VM: 
    
Agar aapko ek complete, separate operating system chahiye ho (jaise ek Linux aur ek Windows system ek hi machine par chalana), toh VM useful hote hain.

#### Docker:
    
Agar aapko lightweight, scalable aur faster application deployment chahiye ho (especially microservices architecture mein), toh Docker ideal hai.

## Portability:

#### VM:
    
VMs ko move karna thoda mushkil ho sakta hai, kyunki har VM ka apna OS aur hardware abstraction layer hota hai.

#### Docker:

Docker containers ko easily move kiya ja sakta hai, kyunki containers ek standardized environment mein run karte hain, jo ek machine se doosre machine par easily transfer ho sakta hai.

#### Summary: 

Agar aapko full OS ki isolation aur virtualization chahiye, toh VM ka use karein. Agar aapko lightweight, fast aur efficient application deployment chahiye ho, toh Docker better option hai.

# DockerFile Kya Hota ha?

Dockerfile ek text file hota hai jo instructions ka sequence define karta hai, jisse Docker container image build hota hai. Is file mein hum ye specify karte hain ki container ko kaise build karna hai, kaunse software install karne hain, kaunse files copy karni hain, aur container ko run karne ka tarika kya hoga.

Dockerfile mein kuch common instructions hoti hain:

FROM: Ye base image specify karta hai jisse hum apna image banaenge. Jaise, FROM ubuntu:20.04.

RUN: Ye command container ke andar kuch operations run karne ke liye use hota hai, jaise software install karna.

COPY: Isse hum local files ko container ke andar copy kar sakte hain.

WORKDIR: Ye container mein ek specific directory set karta hai, jahan pe hum kaam karenge.

CMD: Ye container ke start hone par default command specify karta hai, jo run hoga.

EXPOSE: Ye specify karta hai kaunse ports ko container expose karega.

# Docker Compose

Docker Compose ek tool hai jo Docker containers ko manage karne ke liye use hota hai. Iska main purpose multiple containers ko ek sath configure karna aur run karna hai. Docker Compose aapko ek configuration file (docker-compose.yml) ke zariye define karne ki suvidha deta hai, jisme aap apne containers, networks, aur volumes ko ek sath manage kar sakte ho.

#### Docker Compose ka use kab hota hai?

Jab aapko ek se zyada containers ko ek sath run karna ho.
    
Jab aapko apne applications ko easily scale karna ho.

Jab aapko apne app ke multiple services ko ek configuration file ke through manage karna ho.

#### Docker Compose kaise kaam karta hai?

Aap ek docker-compose.yml file banate ho, jisme aapko containers, networks, aur volumes ka description diya jata hai.
    
Phir, ek simple command docker-compose up ke zariye aap apne saare containers ko start kar sakte ho.

# What is Docker Swarm?

Docker Swarm ek container orchestration tool hai jo Docker containers ko manage karne ke liye use hota hai. Iska main purpose multiple Docker engines ko ek saath connect karke unko manage karna hai, jisse large-scale applications ko efficiently deploy aur scale kiya ja sake.

Jab aap Docker Swarm use karte hain, to aap multiple machines (ya nodes) ko ek single cluster mein convert kar sakte hain, jahan par containers ko automatically distribute kiya jata hai aur unka load balance bhi manage hota hai.

#### Cluster Management: 
    
Docker Swarm multiple Docker hosts ko ek cluster mein combine karta hai. Isme ek Manager node hota hai, jo cluster ko control karta hai, aur Worker nodes hote hain, jo tasks ko execute karte hain.

#### High Availability:

Agar ek node fail ho jata hai, to Swarm automatically uss node par chal rahe containers ko kisi aur healthy node par transfer kar deta hai, taaki application ka uptime maintain ho sake.

#### Scaling: 
    
Aap apne services ko easily scale kar sakte hain. Agar aapko apne application ke containers ki number badhani ho, to Docker Swarm automatically naye containers deploy kar dega.

#### Load Balancing: 
    
Swarm automatically incoming traffic ko services ke across distribute karta hai, taaki load efficiently balance ho jaye.

#### Service Discovery: 

Docker Swarm mein service discovery hoti hai, matlab containers apas mein easily communicate kar sakte hain, chahe wo different nodes par ho.

#### Rolling Updates: 

Aap apne services ko bina downtime ke update kar sakte hain. Swarm apne containers ko gradually update karta hai, jisse application ka availability continuously maintained rahe.
